Clustering vehicles based on their specifications.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
np.random.seed(0)
data_size = 300
data = {
    'weight': np.random.randint(1000, 3000, data_size),
    'Engine Size':np.random.uniform(1.0,4.0,data_size),
    'HorsePower':np.random.randint(50,300,data_size)
}
```

```python
data_df = pd.DataFrame(data)
print(data_df.head())
x=data_df
```

```python
kmeans =  KMeans(n_clusters=3, random_state=42)
kmeans.fit(x)
```

```python
plt.scatter(data_df['weight'],data_df['HorsePower'])
plt.xlabel('Weight')
plt.ylabel('HorsePower')
plt.title("Vehicle Clusters")
plt.show()
```
